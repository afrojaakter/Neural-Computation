```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def weight_bias(w, theta):
  W_1 = np.random.normal(0, w, (n_h, n))  # weight: Input layer to first hidden layer
  b_1 = np.random.normal(0, theta, (n_h, 1))  #bias: first hidden layer
  W_h = np.random.normal(0, w, (n_h, n_h))    # weight: hidden layer to hidden layer
  b_h = np.random.normal(0, theta, (n_h, 1))  #bias: hidden layer   
  W_o = np.random.normal(0, w, (n_o, n_h))    #weight: Last hidden layer to output layer
  b_o = np.random.normal(0, theta, (n_o, 1))  #bias: output layer
  return W_1, b_1, W_h, b_h, W_o, b_o


def network(axs,X,L,trial):
  '''
  param n_h: Number of neurons in the hidden layer
  param L: Number of hidden layers
  param grid: Input Data
  '''
  A = [0.25, 0.5, 1, 2, 4]  #the variance scale
  for weightIndex in range(len(A)):
    w = A[weightIndex]
    W_1, b_1,W_h, b_h, W_o, b_o = weight_bias(weightIndex, weightIndex)
    X_L = np.array([])
    
    for j in range(0,N):
      x = X[j,:].reshape(n,1)
      x_h1 = np.tanh(np.dot(W_1, x) + b_1)
      for i in range(L):
        x_h = np.tanh(np.dot(W_h, x_h1) + b_h)
        x_h1 = x_h

      x_l = np.tanh(np.dot(W_o, x_h) + b_o)
      
      X_L = np.append(X_L, x_l)
    
    axs[trial,weightIndex].scatter(X[:,0],X[:,1],None, X_L, '.',cmap=plt.cm.RdBu )
    axs[trial, weightIndex].set_title('A = {}'.format(A[trial]))
    axs[trial, weightIndex].axis('off')



N = 1000  #Number of samples
n = 2   # Dimension of input data
n_h = 15 # Number of neurons in the hidden layer
n_o = 1 # Number of neurons in the output layer
fig, axs = plt.subplots(5,5, figsize=(10,6))

for trial in range(5):
  X = np.random.uniform(-1, 1, (N,n))  # Input data
  network(axs,X,10,trial)  
```
